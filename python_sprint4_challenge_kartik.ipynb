Write a Python program to simulate an MCQ (multiple choice questions) quiz comprising 10 questions. You can frame your own questions. Every question will have 4 options, and exactly one option will be the correct answer.
To answer a question, the student needs to enter the number corresponding to the chosen option. Any entry besides 1, 2, 3, or 4 will lead to the question marked as zero.
Every correct answer will carry 2 marks. Print the final score to the user.

```python
Q={"Number of Continents on Earth":"1:Seven 2:Four 3:Five 4:Six", 
   "Which country successfully landed a man on moon":"1:USA 2:Germany 3:France 4:England",
    "Which country is not present in the continent of North America":"1:Mexico 2:USA 3:Cannada 4:Brazil",
     "Which is the American National Bird":"1:Parrot 2:Crane 3:Eagle 4:Swan", 
   "Which country shares the longest border with USA": "1:Mexico 2:Brazil 3:Cannada 4:Argentina",
"Which is the financial capital of USA ": "1:Sanfrancisco 2: Boston 3:Los Angeles  4:NYC",
"Which of the following space agencies belongs to the USA": "1:BNSC 2:ISRO 3:DLR 4:NASA",
"Which is the second most spoken language in USA": "1:English 2:Spanish 3: Mandarin 4: Italian",
"Which is the capital city of USA ":"1:NYC 2:Washington DC 3: Austin 4: San Jose",
"Mt Rushmore has the sclupture of ": "1:Former US Presidents 2:Hollywood stars 3:Military Personel 4:Scientists"
  }






print("Welcome to the quiz!\n")
A=list(Q.values())
CA =[1,1,4,3,3,4,4,2,2,1]
myiter = iter(A)
myite=iter(CA)

cor_ans= 0
wrong_ans=0
skipped=0
invalid=0
score = 0

for i in Q:
    print(i)
    a=next(myiter)
    print(a)
    ans = int(input("Enter your choice between 1 and 4: "))
    
    correct_ans=next(myite)
    if ans>=1 and ans<=4:
        if ans==correct_ans:
            score=score+1
            cor_ans=cor_ans+1
        elif ans!=correct_ans:
            score=score-1
            wrong_ans = wrong_ans+1
    elif ans==0:
        score=score
        skipped=skipped+1
    else:
        print("invalid option")
        invalid  =invalid+1
        score=score

print(" ")
print(" ")
print(" ")
print("invalid option: ",invalid)
print("")
print("correct ans: ",cor_ans)
print("")
print("wrong: ",wrong_ans)
print(" ")
print("skipped",skipped)
print(" ")
print("your score:",score)

    
    
```
