```python
def seperate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:
        if df[column].nunique() < 100:
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous

categorical, continuous = seperate_data_types(df)

#Tabulate is a package used to print the list, dict or any data sets in a proper format; in a table format

from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical, "Continuous":continuous}, headers = ["categorical", "continuous"]))
            
```

```python
def numerical_info(data, numerical):
    for col in numerical:
        mean = data[col].mean()
        median = data[col].median()
        mode = data[col].mode()[0]
        std = data[col].std()
        min_val = data[col].min()
        max_val = data[col].max()
        print(f"{col}:")
        print("Mean:", mean)
        print("Median:", median)
        print("Mode:", mode)
        print("Standard Deviation:", std)
        print("Minimum:", min_val)
        print("Maximum:", max_val)
        print()
```

```python
numerical_info(df,["PRICEEACH"])
```

```python
numerical_info(df,["SALES"])
```

```python
numerical_info(df,["MSRP"])
```

```python
def categorical_info(data, categorical):
    for col in categorical:
        mode = data[col].mode()[0]
        min_val = data[col].min()
        max_val = data[col].max()
        unique_vals = data[col].nunique()
        print(f"{col}:")
        print("Mode:", mode)
        print("Minimum:", min_val)
        print("Maximum:", max_val)
        print("Unique Values:", unique_vals)
        print()
```

```python
categorical_info(df,["PRODUCTLINE"])
```

```python
categorical_info(df,["CUSTOMERNAME"])
```

```python
categorical_info(df,["COUNTRY"])
```

# Task 3
Perform descriptive analysis of the numerical columns
'PRICEEACH', 'SALES', and 'MSRP' by using the following:
• Bar chart, Pie chart
Measures of central tendency (mean, median)
Standard deviation of the column
Skewness of the column

```python
def histogram_boxplot(data, numerical):
    for col in numerical:
        plt.figure()
        sns.histplot(data[col])
        plt.title(f"Histogram for {col}")
        plt.xlabel(col)
        plt.ylabel("Frequency")
        plt.show()
        plt.figure()
        sns.boxplot(data[col])
        plt.title(f"Boxplot for {col}")
        plt.xlabel(col)
        plt.ylabel("Value")
        plt.show()
        print(f"{col}:")
        print("Mean:", data[col].mean())
        print("Median:", data[col].median())
        print("Standard Deviation:", data[col].std())
        print("Skewness:", data[col].skew())
        print()
```

```python
histogram_boxplot(df,["PRICEEACH"])
```

```python
histogram_boxplot(df,["SALES"])
```

```python
histogram_boxplot(df,["MSRP"])
```

# Task 4
Perform descriptive analysis of the categorical columns
'PRODUCTLINE', 'CUSTOMERNAME', and
'COUNTRY' by using the following:
• Bar chart, Pie chart
• Measures of central tendency (mean, median)
Standard deviation of the column
• Skewness of the column

```python
def cat_chart(col):
    print("Mode:",df[col].mode()[0])
    print("Missing values :",df[col].isnull().sum())
    dist= df[col].value_counts()
    fig, ax=plt.subplots(1,2, figsize= (15,6))       
    ax[0].set_title("count of"+str(col))           
    labels= list(dist. index)                             

    sns.countplot(x= df[col], ax= ax[0])              
    plt.pie(dist, labels= labels, autopct="%0.2f%%")      

    plt.show()
```

```python
cat_chart("PRODUCTLINE")
```

```python
cat_chart("CUSTOMERNAME")
```

```python
cat_chart("COUNTRY")
```
