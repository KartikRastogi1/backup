Task 4
Calculate and display the average of test scores of students at each school. Display the name of the school that has the highest average test score.
Hint: Use the "group by" operation to first segregate the data for each school and then calculate the averageÂ testÂ score.

```python
avg_test1_scores=dataframe.groupby(by = ["school"])["test1"].mean()
avg_test1_scores
```

```python
print("The name of the school that has the highest average test score.")
avg_test1_scores.sort_values(ascending = False)[:1]
```
