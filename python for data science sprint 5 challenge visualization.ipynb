# Task 4
Display the car mileages for the following four companies using a pivot table:
Hyundai, Mahindra, Renault, and Skoda
Create 4 separate pivot tables capturing the following details:
1). Mean of city mileage
2). Median of city mileage
3). Mean of highway mileage
4). Median of highway mileage

```python
condition = cars[(cars["Make"]=="Hyundai") | (cars["Make"]=="Mahindra") | (cars["Make"]=="Renault") | (cars["Make"]=="Skoda")]
condition
```

```python
print("Mean of city mileage")
pd.pivot_table(condition, index=["Make"], values=["City_Mileage_km_litre"], aggfunc = "mean") 
```

```python
print("Median of city mileage")
pd.pivot_table(condition, index=["Make"], values=["City_Mileage_km_litre"], aggfunc = "median") 
```

```python
print("Mean of highway mileage")
pd.pivot_table(condition, index=["Make"], values=["Highway_Mileage_km_litre"], aggfunc = "mean") 
```

```python
print("Median of highway mileage")
pd.pivot_table(condition, index=["Make"], values=["Highway_Mileage_km_litre"], aggfunc = "median") 
```

# Task 6
Display the average mileage (city and highway) of all manufactures using a horizonal bar chart, in ascending order of mileage.
Note: The missing values in the mileage column need to be filled with the mean value.

```python
cars["City_Mileage_km_litre"].isnull().sum()
```

```python
cars_mean=cars["City_Mileage_km_litre"].mean()
cars_mean
```

```python
cars["City_Mileage_km_litre"].fillna(cars_mean,inplace=True)           #filling missing values with mean
```

```python
cars["City_Mileage_km_litre"].isnull().sum()                               #zero missing values left       
```

```python
cars["Highway_Mileage_km_litre"].isnull().sum()                          #This is for highway mileage
```

```python
b = cars["Highway_Mileage_km_litre"].mean()
b
```

```python
cars["Highway_Mileage_km_litre"].fillna(b,inplace= True)                      #filled with mean values
```

```python
cars["Highway_Mileage_km_litre"].isnull().sum() 
```

```python
x= pd.DataFrame(cars.groupby(by=["Make"])[["City_Mileage_km_litre","Highway_Mileage_km_litre"]].mean())
total_avg = pd.DataFrame(x["City_Mileage_km_litre"] + x["Highway_Mileage_km_litre"].sort_values())
total_avg
```

```python
total_avg.plot(kind="barh",figsize=(10,20))
```
