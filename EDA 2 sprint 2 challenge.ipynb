# Task 1
Define a function to identify the data type of the columns as categorical or numerical.
Separate the names of the categorical and numerical columns and save them into two different lists.

```python
def sep_data(df):
    categorial = []
    continuous = []
    for i in df.columns:
        if df[i].nunique() < 50:
            categorial.append(i)
        else:
            continuous.append(i)
    return categorial, continuous

categorial, continuous = sep_data(df)

from tabulate import tabulate
table = [categorial, continuous]
print(tabulate({"categorial":categorial, "continuous":continuous}, headers=["categorial", "continuous"]))
```

# Task 3
Perform bivariate and multivariate analysis on the numerical columns (identified in Task 1) using appropriate graphs/charts.

```python
def num_num(num1, num2):
    plt.scatter(df[num1], df[num2])
    plt.title(num1+ "vs" + num2 )                 
    plt.xlabel(num1 + "   correlation value = "+str(df[num1].corr(df[num2])))         
    plt.ylabel(num2)
    plt.show()
```

```python
num_num("Age","Hours_Per_Week")
```
