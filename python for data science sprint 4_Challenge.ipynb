Task 2
Thomas, a potential customer from California, is interested in buying a low-priced car, manufactured in 2014. John needs to identify 100 lowest priced cars from the year 2014 with CA as the state of registration. Help John create and display the details of potential cars.

```python
condition = (df["year"]==2014) & (df["state"]=="ca")
x=df[condition].sort_values("mp")[:100]
x
```

Task 3
Thomas is confused looking at the long list of 100 cars. Help John to display the output in a cross table based on the car manufacturers and prices.

```python
pd.pivot_table(x,index = ["make"], values=["mp"], aggfunc = ["mean"])
```

```python
cross_table = pd.crosstab(x['make'], x['mp'], rownames=["make"], colnames=["mp"], values=x['mp'], aggfunc='mean') 
```

Task 4
Thomas is particularly interested in Chevrolet cars. Eric is willing to offer a 10% discount on Chevrolet cars from the list of 100 cars identified in the previous tasks.
Display the updated data with only Chevrolet cars with their original prices and discounted prices.

```python
condition2=(x["make"]=="Chevrolet")
chevrolet  = x[condition2]
chevrolet["Discount price"] = chevrolet["mp"]*0.9
chevrolet[["make","mp", "Discount price"]]
```

Task 5
Thomas and several other customers seem to be interested in the cars with black or gray colored interiors. John wants to figure out if the cars with interiors in black or gray are priced higher or lower than cars having interiors of other colors.
Determine the average price of all the 100 lowest priced cars from the year 2014 based on their interior colors.

```python
condition3=(df["year"]==2014)
```

```python
lowest_prices= df[condition3].sort_values("mp")[:100]
lowest_prices
```

```python
lowest_prices.groupby(by = ["interior"])["mp"].mean()
```

```python
year = df["year"]==2014
df3 = df[year]
print("other interior colors")
df3.groupby(by = ["interior"])["mp"].mean()
```

Task 7
Eric, another customer is interested in a good condition BMW Sedan manufactured between 2010 and 2015. John wants to display the details of all such cars that have a condition score >=4.5.
Help John to figure out of the year-wise maximum expected price (mp) of the above BMW cars manufactured between the year 2010 and 2015 grouped by each model.

```python
year_2010_and_2015 = df[(df["year"].between(2010, 2015, inclusive = "both")) & (df["condition"]>=4.5) & (df["make"]=="BMW") & (df["body"]=="Sedan")]
year_2010_and_2015
```

```python
year_2010_and_2015.groupby(by = ["year"])["mp"].max()
```
